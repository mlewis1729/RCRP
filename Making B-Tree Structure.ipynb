```python
import RCRP as rcrp
import yfinance as yf
import numpy as np, pandas as pd
from datetime import datetime
```

```python
Ticker_Name_File = './data_Files/NYSE.csv'
df = pd.read_csv( Ticker_Name_File )
yearToRun = 2022
start_date, end_date = str( yearToRun - 3 ) + '-01-01', str( yearToRun + 1 ) + '-03-01'
maxCorr = 0.80
```

```python
tickers = df['Symbol'].unique()
ClosePrices = {}
i = 0
t1 = datetime.now()
for ticker in tickers:
    i += 1
    try:
        data = yf.download(ticker, start_date, end_date)
    except:
        continue
    ClosePrices[ticker] = data[data.index >= start_date].Close
    if i % 10 == 0:
        t2 = datetime.now()
        leftover = len(tickers) - i
        rate = (t2-t1).seconds * 1. / i
        print( i, ",", leftover * rate, " seconds to go" )
```

```python
returns = pd.DataFrame(ClosePrices).ffill().pct_change()
UseCompanies  = ( returns.apply(lambda x: np.isnan(x).sum(), axis=0) <= 1 )
UseCompanies  = UseCompanies & ( returns.apply(lambda x: np.abs(x) > 0., axis=0).sum() >= 1 )
returns = returns[returns.columns[UseCompanies]]
CompanySet = returns.columns
StartLoop = True
while StartLoop or len( C[C.abs()>maxCorr] ) > 0:
    StartLoop = False
    cov = returns.cov()
    corr = returns.corr()
    C = corr.stack()
    C = C[C.index.get_level_values(0) < C.index.get_level_values(1) ]
    if len( C[C.abs()>maxCorr] ) > 0:
        CompanySet = CompanySet[CompanySet!=C[C.abs()>maxCorr].index.get_level_values(1)[0] ]
        returns = returns[CompanySet]
while np.min(np.diag(cov)) <= 0.:
    tmp = pd.Series(np.diag(cov), index=cov.index)
    CompanySet = CompanySet[~CompanySet.isin(tmp[tmp<=0.].index)]
    returns = returns[CompanySet]
    cov = returns.cov()
    corr = returns.corr()
```

```python
np.random.seed(123)
t1 = datetime.now()
w     = rcrp.RCRP( cov )
t2 = datetime.now()
runTime = ( t2 - t1 ).seconds
print('Runtime (seconds):', runTime )
```

```python
np.random.seed(123)
t1 = datetime.now()
w2     = rcrp.RCRP( cov, use_extended_terms=False )
t2 = datetime.now()
runTime = ( t2 - t1 ).seconds
print('Runtime (seconds):', runTime )
```

```python
np.random.seed(123)
t1 = datetime.now()
Ct = rcrp.Cov_Tree()
Ct.fit( cov )
t2 = datetime.now()
runTime = ( t2 - t1 ).seconds
print('Runtime (seconds):', runTime )
```

```python
b = None # pd.Series( 1., index = cov.index )
t1 = datetime.now()
w1_test = Ct.solve( b, use_extended_terms=True )
t2 = datetime.now()
runTime = ( t2 - t1 ).seconds
print('Runtime (seconds):', runTime )
t1 = datetime.now()
w2_test = Ct.solve( b, use_extended_terms=False )
t2 = datetime.now()
runTime = ( t2 - t1 ).seconds
print('Runtime (seconds):', runTime )
```
